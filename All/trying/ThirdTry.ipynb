```python
import spacy

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

def identify_subjects_and_limitations(query):
    # Process the query with spaCy
    doc = nlp(query)

    subjects = []
    limitations = []

    # Define common subject keywords (can be expanded as needed)
    subject_keywords = {'watch', 'watches'}

    # Iterate over tokens in the document
    for token in doc:
        # Identify subjects by direct keyword match (expandable with synonyms or similar words)
        if token.lemma_ in subject_keywords:
            subjects.append(token.text)

        # Extracting limitations by checking dependency tags and entity types
        if token.dep_ in ('nummod', 'amod', 'nsubj', 'dobj', 'pobj') or token.ent_type_ in ('MONEY', 'QUANTITY'):
            limitations.append(token.text)
        # Further refine to capture phrases using entity heads and children
        if token.head.ent_type_ == 'MONEY':
            limitations.append(token.head.text + ' ' + token.text)

    return subjects, limitations

# Example queries
queries = [
    "Show me watches under Rs. 500.",
    "Can you list watches between Rs. 1000 and Rs. 2000?",
    "Show me all Samsung watches available.",
]

for query in queries:
    subjects, limitations = identify_subjects_and_limitations(query)
    print(f"Query: {query}\nSubjects: {subjects}\nLimitations: {limitations}\n")

```

```python
import spacy

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

def identify_subjects_and_limitations(query):
    doc = nlp(query)
    subjects = []
    limitations = []

    # We keep a track of tokens that form part of any captured limitation to avoid duplicates
    limitation_tokens = set()

    for token in doc:
        # Identify subjects by direct keyword match
        if token.lemma_ in {'watch', 'watches'}:
            subjects.append(token.text)

        # Check if token is a named entity and of a certain type relevant to limitations
        if token.ent_type_ in ('MONEY', 'QUANTITY', 'ORG'):
            # We append the entity itself
            limitations.append(token.text)
            limitation_tokens.add(token.i)

            # Check for any prepositions or conjuncts linking this entity for a range or list
            head = token.head
            if head.pos_ in ('ADP', 'CONJ') and head.dep_ in ('prep', 'cc'):
                # Capture the whole phrase by looking for the subtree of the head
                phrase = ' '.join([child.text for child in head.subtree if child.i not in limitation_tokens])
                limitations.append(phrase)
                limitation_tokens.update([child.i for child in head.subtree])

    # Eliminate duplicates and cleanup
    limitations = list(set(limitations))  # remove duplicates that might have been added

    return subjects, limitations

# Example queries
queries = [
    "Show me watches under Rs. 500.",
    "Can you list watches between Rs. 1000 and Rs. 2000?",
    "Show me all Samsung watches available.",
    "Which watches do you have from Jianuo?",
    "List watches sold by New-Gen.",
    "Show me the top-rated sellers who sell watches.",
    "Show me watches from sellers with more than 90% positive ratings.",
    "Which watches are sold by highly rated sellers?",
    "I'm looking for sports watches priced between Rs. 1500 and Rs. 2500, sold by top-rated sellers.",
    "What are the latest smartwatches available under Rs. 5000?",
    "List all luxury watches.",
    "Help me find a watch for a gift under Rs. 2000.",
]

for query in queries:
    subjects, limitations = identify_subjects_and_limitations(query)
    print(f"Query: {query}\nSubjects: {subjects}\nLimitations: {limitations}\n")
```

```python
def format_output(subjects, limitations):
    formatted_output = []

    # Handle price-related limitations
    for limitation in limitations:
        if 'Rs' in limitation:
            action, currency = limitation
            if action == 'under' or action == 'over':
                formatted_output.append(f"Price {action} {currency}")
            elif action == 'between':
                # Assuming the next two elements in the list are part of the range
                start_price = currency
                end_price = limitations[limitations.index(limitation) + 2][1]
                formatted_output.append(f"Price {action} {start_price} and {end_price}")
    
    # Handle brand and seller entities
    for subject in subjects:
        if any(x in subject.lower() for x in ['samsung', 'jianuo', 'new-gen']):
            if 'seller' in query.lower() or 'sold by' in query.lower():
                formatted_output.append(f"Seller is {subject}")
            else:
                formatted_output.append(f"Brand is {subject}")
        
        if 'rated sellers' in subject or 'positive ratings' in subject:
            # Check for numeric value associated with ratings
            rating_info = next((lim[1] for lim in limitations if '%' in lim[1]), None)
            if rating_info:
                formatted_output.append(f"Rating is positive ratings of {rating_info}")
            else:
                formatted_output.append(subject)

    return formatted_output

# Example of converting raw output into formatted output
raw_outputs = [
    (['Rs'], [('under', 'Rs'), ('.', '500')]),
    ([], [('between', 'Rs'), ('.', '1000'), ('.', '2000')]),
    (['Samsung', 'available watches'], []),
    (['New-Gen.'], []),
    (['positive ratings'], [('%', '90')])
]

for subjects, limitations in raw_outputs:
    print(format_output(subjects, limitations))
```

```python
def format_output(subjects, limitations):
    formatted_output = []

    # Mapping of price-related actions to human-readable formats
    price_actions = {
        'under': "Price under Rs. {}",
        'over': "Price over Rs. {}",
        'between': "Price between Rs. {} and Rs. {}"
    }

    # Iterate over limitations to extract and format price-related information
    for i, limitation in enumerate(limitations):
        action, value = limitation
        if action in price_actions:
            if action == 'between':
                # Ensure there is enough data following the 'between' keyword to format correctly
                if i + 2 < len(limitations):  # Checks that two more elements exist
                    start_price = limitations[i + 1][1].replace('.', '')  # Cleans up the 'Rs.' and potential extra period
                    end_price = limitations[i + 2][1].replace('.', '')
                    formatted_output.append(price_actions[action].format(start_price, end_price))
            else:
                # Formats for 'under' and 'over'
                formatted_output.append(price_actions[action].format(value.replace('.', '')))

    # Process subjects for brand and seller names
    for subject in subjects:
        if "watches" not in subject.lower():  # Filter out generic mentions of watches
            if 'rated' in subject or '%' in subject:
                # Handle ratings specifically
                formatted_output.append(f"Rating is positive ratings of {subject}")
            elif any(company in subject for company in ["Samsung", "Jianuo", "New-Gen"]):
                # Assume these are brands or sellers
                formatted_output.append(f"Brand is {subject}")

    return formatted_output

# Example outputs from initial spaCy analysis
outputs = [
    (['Rs'], [('under', 'Rs'), ('.', '500')]),
    ([], [('between', 'Rs'), ('.', '1000'), ('.', '2000')]),
    (['Samsung watches', 'available'], []),
    (['New-Gen.'], []),
    (['positive ratings'], [('%', '90')])
]

# Applying the formatting function to each query output
for subjects, limitations in outputs:
    print(format_output(subjects, limitations))
```
